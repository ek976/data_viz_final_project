```python
# highest voted 

import matplotlib.pyplot as plt

# Hex colors
hex_colors = [
    "#3d6469", "#ffa205", "#ff4500", "#08030a", "#feeece", "#d40637", "#5f0922"
]

# Calculate number of rows and columns based on the number of colors
num_colors = len(hex_colors)
num_cols = min(num_colors, 4)
num_rows = (num_colors + num_cols - 1) // num_cols

# Create a figure and axis
fig, ax = plt.subplots(figsize=(num_cols * 2, num_rows * 2))

# Plot each color as a square
for i, color in enumerate(hex_colors):
    ax.add_patch(plt.Rectangle((i % num_cols, i // num_cols), 1, 1, color=color))

    # Add text with hex value
    ax.text((i % num_cols) + 0.5, (i // num_cols) + 0.5, color, color='black', ha='center', va='center')

# Set axis limits and labels
ax.set_xlim(0, num_cols)
ax.set_ylim(0, num_rows)
ax.set_xticks(range(num_cols + 1))
ax.set_yticks(range(num_rows + 1))
ax.set_xticklabels([])
ax.set_yticklabels([])

# Remove axis
ax.axis('off')

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('code/govt_spend_primary.csv')
```

```python
df.head(5)
```

```python
df=df.drop(columns=['country_code', 'series_name'])
df.head(10)
```

```python
df1 = df.pivot(index='country_name', columns='year', values='value')
df1.reset_index(inplace=True)

```

```python
df1.head(10)
```

```python
df1.head(10)
```

```python
missing_data = df1.isnull().sum()
print("Missing Data Column-wise:")
print(missing_data)
```

```python
df1.columns
```

```python
columns_to_check = [2010, 2011, 2012]  
df1.dropna(subset=columns_to_check, inplace=True)

# Display the DataFrame after dropping rows
print(df1)
```

```python
df1.head(10)
```

```python
columns_to_extract = ['country_name', 2010, 2011, 2012]
df2 = df1[columns_to_extract]
print(df2)
```

```python
df2.rename(columns={2010: 'year1', 2011: 'year2', 2012: 'year3'}, inplace=True)
```

```python
df2.head(10)
```

```python
df2.columns
```

```python
df2.to_csv('tidy.csv', index=False) 
```

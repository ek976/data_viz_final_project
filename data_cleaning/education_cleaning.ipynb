```python
import pandas as pd 
```

```python
# read in data 
education_data = pd.read_csv('../data/EdStatsData.csv') 
education_data.head()
```

```python
# Print column names
print("Column Names:")
print(education_data.columns)
```

```python
# Specify the columns to keep
columns_to_keep = ['Country Name', 'Country Code', 'Indicator Name', '2010', '2011', '2012', '2013', '2014']

# Subset the DataFrame to keep only the specified columns
education_data = education_data[columns_to_keep]

# Display the DataFrame after subsetting
print("\nDataFrame after keeping selected columns:")
print(education_data.head())
```

```python
# Print unique values in the 'Indicator Name' column
unique_indicators = education_data['Indicator Name'].unique()

print("\nUnique Indicator Names:")
for indicator in unique_indicators:
    print(indicator)
```

```python
# List of indicator names to keep
indicators_to_keep = [
    'Gross enrolment ratio, lower secondary, female (%)',
    'Gross enrolment ratio, lower secondary, male (%)',
    'Gross enrolment ratio, post-secondary non-tertiary, female (%)',
    'Gross enrolment ratio, post-secondary non-tertiary, male (%)',
    'Gross enrolment ratio, pre-primary, female (%)',
    'Gross enrolment ratio, pre-primary, male (%)',
    'Gross enrolment ratio, primary and lower secondary, female (%)',
    'Gross enrolment ratio, primary and lower secondary, male (%)',
    'Gross enrolment ratio, primary and secondary, female (%)',
    'Gross enrolment ratio, primary and secondary, male (%)',
    'Gross enrolment ratio, primary to tertiary, female (%)',
    'Gross enrolment ratio, primary to tertiary, male (%)',
    'Gross enrolment ratio, primary, female (%)',
    'Gross enrolment ratio, primary, male (%)',
    'Gross enrolment ratio, secondary, female (%)',
    'Gross enrolment ratio, secondary, male (%)',
    'Gross enrolment ratio, tertiary, female (%)',
    'Gross enrolment ratio, tertiary, male (%)',
    'Gross enrolment ratio, upper secondary, female (%)',
    'Gross enrolment ratio, upper secondary, male (%)',
    'GNI (current US$)',
    'GNI per capita, Atlas method (current US$)',
    'GDP per capita (current US$)',
    'Government expenditure in educational institutions as % of GDP (%)'
]

# Filter the DataFrame to keep only the specified indicator names
filtered_education_data = education_data[education_data['Indicator Name'].isin(indicators_to_keep)]

# Display the filtered DataFrame
print("\nDataFrame after filtering for specific indicators:")
print(filtered_education_data.head())
```

```python
# file path
output_file_path = '../cleaned_data/filtered_education_data.csv'

# Save the filtered DataFrame to a CSV file
filtered_education_data.to_csv(output_file_path, index=False)

print("Filtered data saved to", output_file_path)
```
